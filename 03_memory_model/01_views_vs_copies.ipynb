```python
import numpy as np
```

```python
x = np.arange(6)
print("x : ",x)
print("x shape : ",x.shape)
print("ndim of x : ",x.ndim)
```

```python
y = x.reshape(2,3)
print("y : ",y)
print("y shape : ",y.shape)
print("ndim of y : ",y.ndim)
```

```python
y[0,0] = 999
print("After changing y[0,0] = 999 : \n",y)
print("Printing x after changing y[0,0] = 999 : \n", x)
```

```python
print("x.base : ", x.base)
print("y.base is x : ", y.base is x)
```

```python
z = x.reshape(2, 3).copy()
z[0, 0] = -1
print("After changing z[0,0] = -1")
print("z :\n", z)
print("y :\n", x)
print("x :\n", x)
```

View = shares memory (changing one changes the other)

Copy = owns memory (changes donâ€™t leak)

reshape often returns a view; .copy() always makes a copy
